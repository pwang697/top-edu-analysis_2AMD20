## Data Preparation

Read datasets

```python
import pandas as pd
import os

wur_folder_path = '.\\datasets\\world_university_rankings'
csv_files = [os.path.join(wur_folder_path, f) for f in os.listdir(wur_folder_path) if f.endswith('.csv')]
wur_df = pd.DataFrame()
for file in csv_files:
    data = pd.read_csv(file)
    data.insert(loc=0,column='year',value=file.split('\\world_university_rankings\\')[1].split('_')[0])
    wur_df = pd.concat([wur_df, data], ignore_index=True)

wur_df.head()
```
