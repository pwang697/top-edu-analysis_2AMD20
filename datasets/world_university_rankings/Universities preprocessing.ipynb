```python
# Load the data from csv file
df_unis = pd.read_csv('Colleges_and_Universities.csv')

# Lowercase column names
df_unis.columns = df_unis.columns.str.lower()

# Drop unnecessary columns
df_unis.drop(df_unis.columns[:4], axis=1, inplace=True)
df_unis.drop(df_unis.columns[6:11], axis=1, inplace=True)
df_unis.drop(df_unis.columns[2], axis=1, inplace=True)
df_unis.drop(df_unis.columns[6:8], axis=1, inplace=True)
df_unis.drop(df_unis.columns[8:], axis=1, inplace=True)
df_unis.drop(df_unis.columns[1], axis=1, inplace=True)
df_unis.head()

# Save the data to csv file
# df_unis.to_csv('Colleges_and_Universities_cleaned.csv', index=False)
```
